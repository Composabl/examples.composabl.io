```python
import os 
import pandas as pd
import matplotlib.pyplot as plt
```

# Read History File

```python
df = pd.read_pickle(f"./history/db.pkl")
```

# Plot Metrics

```python
plt.figure(figsize=(10,5))
plt.subplot(3,1,1)
plt.scatter(df.reset_index()['time'],df.reset_index()['reward'],s=0.5, alpha=0.2)
plt.ylabel('Reward')
plt.legend(['reward'],loc='best')

plt.subplot(3,1,2)
plt.scatter(df.reset_index()['time'],df.reset_index()['balance'],s=0.5, alpha=0.1)
plt.ylabel('Balance $')
plt.legend(['Balance'],loc='best')

plt.subplot(3,1,3)
plt.scatter(df.reset_index()['time'],df.reset_index()['order_cutoff'],s=0.6, alpha=0.1)
#plt.scatter(df.reset_index()['time'],df.reset_index()['Cref'],s=0.6, alpha=0.9)
plt.ylabel('Order Cutoff')
plt.legend(['Order Cutoff'],loc='best')
plt.xlabel('iteration')
```
