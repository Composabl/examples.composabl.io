```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_pickle('data/mpc_history3.pkl')
```

```python
plt.figure(figsize=(15,10))
plt.subplot(4,1,1)
#plt.plot(self.reward_history, 'r.-')
plt.scatter(df.reset_index()['time'],df.reset_index()['reward'],s=0.5, alpha=0.3)
plt.ylabel('Reward')
plt.legend(['reward'],loc='best')

plt.subplot(4,1,2)
#plt.plot(self.rms_history, 'r.-')
plt.scatter(df.reset_index()['time'],df.reset_index()['x'],s=0.5, alpha=0.1)
plt.scatter(df.reset_index()['time'],df.reset_index()['y'],s=0.5, alpha=0.1)
plt.axhline(0, color='r')
plt.ylabel('X and Y')
plt.legend(['x', 'y','target'],loc='best')

plt.subplot(4,1,3)
plt.scatter(df.reset_index()['time'],df.reset_index()['angle'],s=0.6, alpha=0.1)
plt.axhline(0, color='r')
plt.ylabel('angle')
plt.legend(['angle', 'target'],loc='best')
plt.xlabel('iteration')

plt.subplot(4,1,4)
ma = df.reset_index()['reward'].rolling(window=400).mean()
plt.plot(ma, 'r')
#plt.plot(df.reset_index()['reward'].rolling(window=400).max(), 'r')
```
