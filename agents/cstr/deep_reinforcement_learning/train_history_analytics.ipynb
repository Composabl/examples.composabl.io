```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Read History File

```python
df = pd.read_pickle('history.pkl')
```

# Plot Metrics

```python
plt.figure(figsize=(10,5))
plt.subplot(3,1,1)
plt.scatter(df.reset_index()['time'],df.reset_index()['reward'],s=0.5, alpha=0.2)
plt.ylabel('Reward')
plt.legend(['reward'],loc='best')

plt.subplot(3,1,2)
plt.scatter(df.reset_index()['time'],df.reset_index()['rms'],s=0.5, alpha=0.1)
plt.ylabel('RMS error')
plt.legend(['RMS'],loc='best')

plt.subplot(3,1,3)
plt.scatter(df.reset_index()['time'],df.reset_index()['Ca'],s=0.6, alpha=0.1)
plt.scatter(df.reset_index()['time'],df.reset_index()['Cref'],s=0.6, alpha=0.9)
plt.ylabel('Ca')
plt.legend(['Ca'],loc='best')
plt.xlabel('iteration')
```

```python
plt.figure(figsize=(15,10))
plt.subplot(4,1,1)
plt.plot(df.reset_index()['reward'].rolling(window=90).mean(), 'b')
plt.plot(df.reset_index()['reward'].rolling(window=400).mean(), 'r')
plt.ylabel('Reward Moving Average')

plt.subplot(4,1,2)

plt.plot(df.reset_index()['time'])
plt.ylabel('Episode Length')

```
