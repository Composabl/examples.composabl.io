```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Read History Files

```python
df_ss1 = pd.read_pickle('ss1_history.pkl')
df_ss2 = pd.read_pickle('ss2_history.pkl')
df_transition = pd.read_pickle('transition_history.pkl')
df_selector = pd.read_pickle('selector_history.pkl')
```

# Plot Metrics

```python
def plot_metrics(df, title):
    plt.figure(figsize=(10,5))
    plt.subplot(3,1,1)
    #plt.plot(self.reward_history, 'r.-')
    plt.scatter(df.reset_index()['time'],df.reset_index()['reward'],s=0.5, alpha=0.2)
    plt.ylabel('Reward')
    plt.legend(['reward'],loc='best')
    plt.title(title)

    plt.subplot(3,1,2)
    #plt.plot(self.rms_history, 'r.-')
    plt.scatter(df.reset_index()['time'],df.reset_index()['rms'],s=0.5, alpha=0.1)
    plt.ylabel('RMS error')
    plt.legend(['RMS'],loc='best')

    plt.subplot(3,1,3)
    plt.scatter(df.reset_index()['time'],df.reset_index()['Ca'],s=0.6, alpha=0.1)
    plt.scatter(df.reset_index()['time'],df.reset_index()['Cref'],s=0.6, alpha=0.1)
    plt.ylabel('Ca')
    plt.legend(['Ca'],loc='best')
    plt.xlabel('iteration')
    
```

```python
#SS1
plot_metrics(df_ss1, 'SS1 Metrics')
```

```python
#SS2
plot_metrics(df_ss2, 'SS2 Metrics')
```

```python
#Transition
plot_metrics(df_transition, 'Transition Metrics')
```

```python
#Selector
plot_metrics(df_selector, 'Selector Metrics')
```
